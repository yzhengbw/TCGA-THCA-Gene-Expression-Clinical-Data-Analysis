# 1.Download clinical and gene expression data

```r
# import libraries
install.packages("ggplot2")
install.packages("pheatmap")
install.packages("ggrepel")
library(ggplot2)
library(pheatmap)
library(dplyr)
library(ggrepel)
library(RColorBrewer)
```

```r
# import thyroid carcinoma data and view
gene_data <- read.csv('/content/Human__TCGA_THCA__UNC__RNAseq__HiSeq_RNA__01_28_2016__BI__Gene__Firehose_RSEM_log2.cct', sep = "\t", header = TRUE)
clinical_data <- read.csv('/content/Human__TCGA_THCA__MS__Clinical__Clinical__01_28_2016__BI__Clinical__Firehose.tsi.txt', sep = '\t', header = TRUE)
View(gene_data)
View(clinical_data)
```

# 2.Check and remove missing values/duplicates of the parameter if any.

### (1) Check duplicates

```r
# Overallsurvival and overall_survival are the same, delete one
clinical_data <- clinical_data[-nrow(clinical_data),]
View(clinical_data)
```

```r
# Check duplicated values
any(duplicated(gene_data))
any(duplicated(clinical_data[,-1]))
```

### (2) Check missing values

```r
na_count = sum(is.na(gene_data))
cat("Total RNA NA values:", na_count, "\n")
na_count = sum(is.na(clinical_data))
cat("Total Clinical NA values:", na_count, "\n")
```

### (3) Combine two dataframes

```r
# Get column names
rna_columns <- colnames(gene_data)
clinical_columns <- colnames(clinical_data)
# Find common columns
common_columns <- intersect(rna_columns, clinical_columns)
# Count the number of common columns
print(length(common_columns))
```

```r
# Keep only common columns
RNA_common <- gene_data[, common_columns, drop = FALSE]
clinical_common <- clinical_data[, common_columns, drop = FALSE]
```

```r
# Merge and transform data
merged_data <- rbind(RNA_common, clinical_common)
merged_data <- t(merged_data)

# Set first row as column names
colnames(merged_data) <- as.character(merged_data[1, ])
# Remove the first row from the data frame
merged_data <- merged_data[-1, ]
str(merged_data)
merged_data <- as.data.frame(merged_data)

# Show
dim(merged_data)
View(merged_data)
```

### (4) Handle missing values

```r
# Check any missing values
na_count <- sum(is.na(merged_data))
cat('Missing values:',na_count,'\n')
```

```r
# Start cleaning, replace all NA with mode

# define fuction
get_mode_str <- function(x) {
  uniq_x <- unique(x)
  uniq_x[which.max(tabulate(match(x, uniq_x)))]
}

# Replace
merged_clean <- merged_data
for (i in 1:ncol(merged_data)) {
  mode_value <- get_mode_str(merged_data[[i]])
  merged_clean[[i]] <- ifelse(is.na(merged_data[[i]]), mode_value, merged_data[[i]])
}
View(merged_clean)
```

```r
na_count <- sum(is.na(merged_clean))
cat('Missing values:',na_count,'\n')
```

```r
#Added after submission, to store cleaned csv
write.csv(merged_clean, file = "cleaned_THCA.csv")
```

# 3.Apply exploration data analysis to your dataset and visualize types of clinical information

### (1) Pie chart for race

```r
# Count the race frequency
race_freq <- merged_clean %>%
  group_by(race) %>%
  summarise(value = n())

# Count percentage
race_freq <- race_freq %>%
  mutate(percentage = value / sum(value) * 100)

# Get position
race_freq <- race_freq %>%
  mutate(csum = rev(cumsum(rev(value))),
  pos = value/2 + lead(csum, 1),
  pos = if_else(is.na(pos), value/2, pos))

# Change race names to more readable ones
race_freq <- race_freq %>%
  mutate(race = recode(race,
  "americanindianoralaskanative" = "American India or Alaska Native",
  "blackorafricanamerican" = "Black or African American",
  "asian" = "Asian",
  "white" = "White"))

# Plot
ggplot(race_freq, aes(x="", y=value, fill=race)) +
  geom_bar(stat="identity", width=1) +
  coord_polar("y") +
  labs(title = "Race Distribution of Thyroid Carcinoma") +
  geom_label_repel(data = race_freq,
    aes(y = pos, label = paste0(value, "%")),
    size = 4.5, nudge_x = 1, show.legend = FALSE) + # add percentage label
  guides(fill = guide_legend(title = "Group")) +
  theme_void() # delete axis and background
```

### (2) Histogram for age data

```r
# Plot age data histogram
age_data <- as.numeric(merged_clean$years_to_birth)
ggplot(merged_clean, aes(x = age_data)) +
  geom_histogram(fill = "#69b3a2", color = "black") +  # Adjust bins if necessary
  labs(title = "Age distribution", x = "Age", y = "Frequency") +
  theme_minimal() # Use a minimal theme
```

# 4.Visualize the distribution of gene expression data.(1 point)

### (1) Heatmap for gene expression

```r
# Heat map for gene expression
# Since the dataset is too large with 19927 genes, the heatmap may not make any sense, I will visualize part of gene
# Prepare data

#Extract part of data
gene_expr_data <- merged_clean[1:200,0:20]
cli_data <- merged_clean[, c("gender", "radiation_therapy", "status", "overall_survival")][1:200,]

# transform data to required format
gene_expr_data <- as.matrix(gene_expr_data)
gene_expr_data <- apply(gene_expr_data, 2, as.numeric)

cli_data$gender <- as.factor(cli_data$gender)  # Ensure categorical columns are factors
cli_data$radiation_therapy <- as.factor(cli_data$radiation_therapy)
cli_data$status <- as.factor(cli_data$status)  # Make sure it's a factor for categorical data
cli_data$overall_survival <- as.numeric(cli_data$overall_survival)

colnames(cli_data) = c("Gender", "Radiation therapy", "Status", "Overall survival")
```

```r
# Validate data
table(is.na(cli_data))
table(is.nan(unlist(cli_data)))
table(is.na(gene_expr_data))
table(is.nan(unlist(gene_expr_data)))
```

```r
# Plot heatmap
row.names(gene_expr_data) <- row.names(merged_clean)[1:200]
row.names(cli_data) <- row.names(merged_clean)[1:200]

heatmap_colors <- colorRampPalette(rev(brewer.pal(n = 7, name ="RdYlBu")))(100)
pheatmap(t(gene_expr_data),
         color = heatmap_colors,
         scale = "row",
         cluster_rows = TRUE,
         cluster_cols = TRUE,
         show_colnames = FALSE,
         cellheight = 15,
         annotation_col = cli_data)
```

### (2) Correlational heatmap

```r
# Prepare data
cli_data$Gender <- as.numeric(factor(cli_data$Gender))
cli_data$`Radiation therapy` <- as.numeric(factor(cli_data$`Radiation therapy`))
cli_data$Status <- as.numeric(factor(cli_data$Status))
cli_data$`Overall survival` <- as.numeric(cli_data$`Overall survival`)
combined_data <- cbind(gene_expr_data, cli_data)

cor_matrix <- cor(combined_data, use = "pairwise.complete.obs")
print(cor_matrix)
```

```r
# Plot map
pheatmap(cor_matrix, main = "Correlation Heatmap of Gene Expression and Clinical Features")
```
